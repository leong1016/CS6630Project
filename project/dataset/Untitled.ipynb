```python
import numpy as np
import pandas as pd
```

```python
test = pd.read_csv('Summary.csv')
test.head()
```

```python
test['TotalDly'] = test.iloc[:,1:7].sum(axis=1)
test['PercentCarrierDly'] = test['TotalCarrierDly'] / test['TotalDly']
test['PercentWeatherDly'] = test['TotalWeatherDly'] / test['TotalDly']
test['PercentNASDly'] = test['TotalNASDly'] / test['TotalDly']
test['PercentSecurityDly'] = test['TotalSecurityDly'] / test['TotalDly']
test['PercentLateAircraftDly'] = test['TotalLateAircraftDly'] / test['TotalDly']
test.head()
```

```python
test.to_csv("Summary2.csv")
```
